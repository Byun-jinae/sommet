```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from bs4 import BeautifulSoup as bs
from selenium import webdriver as wb
from selenium.webdriver.common.keys import Keys

import requests as req
from tqdm import tqdm_notebook
import time

from sklearn.preprocessing import StandardScaler
```

```python
# 매일 전체종목 파일에 마지막 행 추가
#마지막 행 생긴거 확인하고 나서!
```

```python
ex_df = pd.read_csv('krx_base.csv',encoding='euc-kr',index_col=0)
ex_df['종목코드'] = ex_df['종목코드'].astype(str).str.zfill(6)
driver = wb.Chrome()
price=[]
trade=[]
for code in ex_df['종목코드']:
    file='날짜전체종목/종목코드-'+str(code)+'.csv'
    list = []
    url = 'https://finance.naver.com/item/frgn.nhn?code='+str(code)+'&page=1'
    driver.get(url)
    html = driver.page_source
    soup = bs(html,'html.parser')
    for i in soup.select('.tlline2 + .type2  .tah'):
        list.append(i.text.strip())
    head = ['날짜', '종가', '전일비', '등락률', '거래량', '기관순매매량', '외국인순매매량', '외국인보유주수', '외국인보유율']
    list = list[:27]
    df2 = np.array(list).reshape(-1,9)
    df2 = pd.DataFrame(df2,columns=head)
    df2 = df2[::-1]
    price.append(pd.to_numeric(df2['종가'][0].replace(',','')))
    trade.append(pd.to_numeric(df2['거래량'][0].replace(',','')))
    rf = pd.read_csv(file,encoding='euc-kr',engine='python',index_col=0)
    rf = rf.append(df2,ignore_index=True)
    rf = rf.drop_duplicates(['날짜'], keep = 'last')
    rf.to_csv(file,encoding='euc-kr')
    print(file)
```

```python
#result 파일을 만들 때 : result는 krx_base를 copy()후  
#for문 끝까지 돌면서 종목코드에 맞는 종목파일 열어서 
#마지막 행의 종가, 거래량, 기관순매매량을 result 컬럼에 추가한다.
#이후 새로운 컬럼을 만들어 필터링 하는 작업은 result 파일로.
#필터링 한 것들은 각각 필터링1:'',필터링2:'' 파일에 저장
```

```python
result = pd.read_csv('krx_base.csv',encoding='euc-kr',index_col=0)
```

```python
result = result[['종목코드','종목명','상장주식수','주주보유주식']]
result['종가'] = price
result['거래량'] = trade
result['시가총액'] = result['상장주식수'] * result['종가']
result['유통금액'] = (result['상장주식수'] - result['주주보유주식']) * result['종가']
```

```python
result = result[result['시가총액']>=150000000000]
result = result[result['종가']>2000]
```

```python
#기관순매매액
```

```python
date_list = [] #시작날짜(누적매수금액0이상)
count_list = [] #횟수(시작날짜부터 매수한 날짜만 세기)

trade_mean = [] #한달평균거래량
sum_list = [] #누적매수금액

result['종목코드'] = result['종목코드'].astype(str).str.zfill(6)
for code in result['종목코드'].values:
    list = []
    df2 = pd.read_csv('날짜전체종목/종목코드-'+str(code)+'.csv',encoding='euc-kr',index_col=0,engine='python')
    df2['기관순매매액'] = pd.to_numeric(df2['기관순매매량'].str.replace(',','')) * pd.to_numeric(df2['종가'].str.replace(',',''))
    temp=0
    check=''
    count=0
    for i in range(len(df2)):
        temp = df2['기관순매매액'].iloc[i]+temp
        if temp<=0:
            temp = 0
            list.append(temp)
            count = 0
            check = ''
        else:
            list.append(temp)
            if check == '':
                check = df2['날짜'].iloc[i]
                count+=1
            else:
                count+=1
    df2['누적매수금액'] = list
    date_list.append(check)
    count_list.append(count)
    sum_list.append(df2['누적매수금액'].iloc[-1])
    df2.to_csv('날짜전체종목/종목코드-'+str(code)+'.csv',encoding='euc-kr')
    trade_mean.append(pd.to_numeric(df2['거래량'].iloc[-5:].str.replace(',','')).mean()) #5일평균거래량평균
    print(code)
```

```python
result['시작날짜'] = date_list
result['누적횟수'] = count_list
result['일주일평균거래량'] = trade_mean
result['누적매수금액'] = sum_list
```

```python
result.to_csv('result.csv',encoding='euc-kr')
result_dropna = pd.read_csv('result.csv',encoding='euc-kr',index_col=0).dropna()
result_dropna.to_csv('result_dropna.csv',encoding='euc-kr')
```

```python
corr_list1 = [] #한달상관관계
corr_list2 = [] #두달상관관계
corr_list3 = [] #시작날짜상관관계
money = [] #기관마지막매매액
from sklearn.preprocessing import StandardScaler

for i in range(len(result_dropna)):
    file = '날짜전체종목/종목코드-'+str(result_dropna['종목코드'].iloc[i])+'.csv'
    data = pd.read_csv(file,encoding='euc-kr',engine='python')
    try:
        data['종가'] = pd.to_numeric(data['종가'].str.replace(',',''))
    except:
        data['종가'] = pd.to_numeric(data['종가'].replace(',',''))
    money.append(data['기관순매매액'].iloc[-1])
    data1 = data[data['날짜']>='2021.01.01'] # 한달
    data2 = data[data['날짜']>='2021.02.01'] # 두달
    data3 = data[data['날짜']>=result_dropna['시작날짜'].iloc[i]] # 두달
    print(i,file)
    
    data_final1=data1[['종가','누적매수금액']]
    data_final2=data1[['종가','누적매수금액']]
    data_final3=data1[['종가','누적매수금액']]
    
    data_scaler = StandardScaler()   
    data_scaler1 = data_scaler.fit_transform(data_final1)
    data_scaler2 = data_scaler.fit_transform(data_final2)
    data_scaler3 = data_scaler.fit_transform(data_final3)
    
    corr_list1.append(pd.DataFrame(data_scaler1).corr()[1][0])
    corr_list2.append(pd.DataFrame(data_scaler2).corr()[1][0])  
    corr_list3.append(pd.DataFrame(data_scaler3).corr()[1][0])
len(corr_list1)
len(corr_list2)
len(corr_list3)
```

```python
result_dropna['한달상관관계'] = corr_list1
result_dropna['두달상관관계'] = corr_list2
result_dropna['시작날짜상관관계'] = corr_list3
result_dropna['마지막기관매매액'] = money

result_dropna.to_csv('result_dropna.csv',encoding='euc-kr')
```

```python
result_dropna['유통금액대비누적매수금액'] = result_dropna['누적매수금액']/(result_dropna['유통금액'])
result_dropna['횟수당매수금액'] = result_dropna['누적매수금액']/result_dropna['누적횟수']
result_dropna['유통금액대비마지막기관매매'] = result_dropna['마지막기관매매액']/result_dropna['유통금액']
result_dropna['ch'] = result_dropna['유통금액']*0.001
```

```python
ub = []
for code in result_dropna['종목코드'].values:    
    file='날짜전체종목/종목코드-'+str(code)+'.csv'
    ex = pd.read_csv(file,encoding='euc-kr',index_col=0,engine='python')
    try:
        ex['종가'] = pd.to_numeric(ex['종가'].str.replace(',',''))    
    except:
        ex['종가'] = pd.to_numeric(ex['종가'].replace(',','')) 
    ex['이평선'] = ex['종가'].rolling(60).mean()
    ex['표준편차'] = ex['종가'].rolling(60).std()
    ex['상단볼린저'] = ex['이평선']+(ex['표준편차']*2)
    ex['하단볼린저'] = ex['이평선']-(ex['표준편차']*2)
    ub.append(ex['상단볼린저'].iloc[-1])    
result_dropna['볼린저상단'] = ub

tf = []
eck=[]
for i in range(len(result_dropna)):
    if result_dropna['종가'].iloc[i]<result_dropna['볼린저상단'].iloc[i]:
        tf.append(0)
    else:
        tf.append(1)
name = []
for i in range(len(result_dropna)):
    if result_dropna['마지막기관매매액'].iloc[i]<0:
        if result_dropna['ch'].iloc[i]<=abs(result_dropna['마지막기관매매액'].iloc[i]):
        #따로 이 종목들만 저정해서 매도종목 알려줄 수 있음
            name.append(result_dropna['종목명'].iloc[i])
            eck.append(0)
        else:
            eck.append(1)
    else:
        eck.append(1)
result_dropna['볼린저상단넘었는지'] = tf
result_dropna['유통주식금액*0.001>매매액'] = eck
```
